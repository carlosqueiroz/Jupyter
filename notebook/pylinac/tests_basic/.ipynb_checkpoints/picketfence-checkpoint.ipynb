```python
from __future__ import print_function
import pydicom
import numpy as np
import matplotlib.pyplot as plt
import sys, os
import glob
import SimpleITK as sitk
import urllib.request
import os.path
import tempfile
import datetime
import urllib.request
from pydicom.dataset import Dataset, FileDataset
from pydicom.uid import ExplicitVRLittleEndian
import pydicom._storage_sopclass_uids
import threading
from pylinac import PicketFence
from myshow import myshow
from pydicom.datadict import dictionary_VR
```

```python
#RTIMAGE TEMPLATE ESTA IMAGEM CONTEM AS TAGS DICOM NECESS√ÅRIAS...
template_img = "test_files/PicketFence/PF_ELEKTA.dcm"  ## EXEMPLO do PYLINAC "test_files/PicketFence/PF.dcm"
# IMAGEM CR
cr_img = 'test_files/PicketFence/RI.dcm'
```

```python
#FUN√á√ÇO QUE INSERE O PIXELDATA DO CR NO PIXELDATA DO RTIMAGE (TEMPLATE) O QUAL CONTEM AS TAGS DICOM "CORRETAS".
#RECEBE TEMPLATE,PIXEL DATA DO CR E O LOCAL DE SALVAMENTO DO DICOM RESULTANTE.
def savedcm(template_dcm,img2,OUTPUT_FILE):
    #EDITANDO O DICOM
    # AQUI EDITA OS DADOS DO PACIENTE E TALS...  COLOCA OS SOPUID ETC...
    # NO RTIMAGE TEM OS DADOS REFERENTES A DISTANCIAS DA FONTE, ISSO TAMB√âM ACREDITO QUE DEVA SER ALTERADO, CONFIRMAR (POIS SE ELE USA FOURIER N√ÉO INTERESSA MUITO PQ SERIA COM A IMAGEM DA TRANSFORMADA QUE INDEPENDE DO TAMANHO E DISTANCIA DO OBJET0...!!
    # ACREDITO QUE O IDEAL SERIA USAR O "TEMPLATE" UM RTIMAGE GERADO NO TPS PRA ESSE TESTE... DAI PEGARIAMOS ESSE RTIMAGE
    # E COM AS INFORMA√á√ïES DELE FARIAMOS S√ì A SUBSTITUI√á√ÉO DO PIXEL DATA DO CR NO LUGAR DO PIXELDATA DESSE RTIMAGE
    #https://pydicom.github.io/pydicom/stable/old/base_element.html
    ###
    
    ## LINHAS E COLUNAS DA IMAGEM REESCALADA
    template_dcm.Rows, template_dcm.Columns = img2.shape
    
    # VALIDA O DICOM https://pydicom.github.io/pydicom/dev/reference/generated/pydicom.dataset.validate_file_meta.html
    pydicom.dataset.validate_file_meta(template_dcm.file_meta, enforce_standard=True)
    
    # PROFUNDIDADE DA IMAGEM http://insightsoftwareconsortium.github.io/SimpleITK-Notebooks/Python_html/01_Image_Basics.html
    image2d = img2.astype(np.uint16)
    
    #CONVERTENDO ARRAY NUMPY PARA BINARIO DO PIXELDATA DO DICOM
    template_dcm.PixelData = image2d.tobytes()
    
    #SALVANDO ARQUIVO NO DISCO
    template_dcm.save_as(OUTPUT_FILE)

#FUN√á√ÉO PARA FAZER A AN√ÅLISE DO picketfence
def readPicketFence(img_dcm):
    #LENDO IMAGEM DICOM QUE FOI REESCALADA
    pf = PicketFence(img_dcm)
    #ANALISANDO
    pf.analyze()
    #GERANDO OS RESULTADOS
    print(pf.results())
    #GERANDO E MOSTRANDO NA TELA O RESULTADO DA IMAGEM FINAL...
    pf.save_analyzed_image(img_dcm+'.png')
    pf.publish_pdf(filename=img_dcm+'.pdf')

    
#Lendo DICOM no PYDICOM
ds = pydicom.dcmread(template_img) #IMAGEM DE TEMPLATE
cr = pydicom.dcmread(cr_img) # IMAGEM CR
cr_pixel_data = cr.pixel_array # D

## CORTANDO IMAGEM ORIGINAL https://stackoverflow.com/questions/55309457/how-to-divide-in-half-pydicom-files-image-using-python
# CR REESCALADO/CORTADO... TEM QUE CORTAR AS BORDAS DO PLATE, IMAGINO EU.... OS RUIDOS DA BORDA DO CR DEVE INFLUENCIAR NA AN√ÅLISE DA BIBLIOTECA... TEM QUE TER MAIS IMAGENS PARA TESTAR...
#cr_re =  cr_pixel_data[180:, 180:]
cr_re =  cr_pixel_data[10:, 20:]


#PLOTANDO OS TAMANHOS DA IMAGEM 
print('Imagem TEMPLATE {} x {} pixels'.format(ds.pixel_array.shape[0],
                                            ds.pixel_array.shape[1]))

print('Imagem CR ORIGINAL {} x {} pixels'.format(cr_pixel_data.shape[0],
                                            cr_pixel_data.shape[1]))

print('Imagem CR RESAMPLED {} x {} pixels'.format(
    cr_re.shape[0], cr_re.shape[1]))
 

OUTPUT_FILE = 'test_files/PicketFence/PFOUT.dcm'

#CHAMANDO FUN√á√ÉO PARA GERAR O DICOM CORRETO
savedcm(ds,cr_re,OUTPUT_FILE)

#CHAMANDO FUN√á√ÉO PARA LER O A IMAGEM PICKET
readPicketFence(OUTPUT_FILE)

```
