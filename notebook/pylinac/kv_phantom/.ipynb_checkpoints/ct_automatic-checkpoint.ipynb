```python
#! -*- encoding:utf-8 -*-
import os
import subprocess
import logging
from time import time
from multiprocessing.pool import ThreadPool
from pprint import pprint
import collections
import shutil

import json
import pydicom
import numpy as np

import pydicom
import dicom
import pprint
import logging

import numpy as np
import matplotlib.pyplot as plt
import sys, os
import glob
import urllib.request
import os.path
import tempfile
import datetime
import urllib.request
import SimpleITK as sitk
from pydicom.dataset import Dataset, FileDataset
from pydicom.uid import ExplicitVRLittleEndian
import pydicom._storage_sopclass_uids
"""
 
from scipy.spatial import distance as dist
from skimage import exposure

# skimage image processing packages
from skimage import measure, morphology
from skimage.morphology import ball, binary_closing
from skimage.measure import label, regionprops
# scipy linear algebra functions 
from scipy.linalg import norm
import scipy.ndimage


"""
import pylab

#https://medium.com/@hengloose/a-comprehensive-starter-guide-to-visualizing-and-analyzing-dicom-images-in-python-7a8430fcb7ed

'''
    For the given path, get the List of all files in the directory tree 
    https://thispointer.com/python-how-to-get-list-of-files-in-directory-and-sub-directories/
'''
def getListOfFiles(dirName):
    # create a list of file and sub directories 
    # names in the given directory 
    listOfFile = os.listdir(dirName)
    allFiles = list()
    # Iterate over all the entries
    for entry in listOfFile:
        # Create full path
        fullPath = os.path.join(dirName, entry)
        # If entry is a directory then get the list of files in this directory 
        if os.path.isdir(fullPath):
            allFiles = allFiles + getListOfFiles(fullPath)
        else:
            allFiles.append(fullPath)
                
    return allFiles   

# Loop over the image files and store everything into a list 
def load_scan(path):    
    # Get the list of all files in directory tree at given path
    listOfFiles = getListOfFiles(path)
  

    # Get the list of all files in directory tree at given path
    listOfFiles = list()
    for (dirpath, dirnames, filenames) in os.walk(path):
        listOfFiles += [os.path.join(dirpath, file) for file in filenames]
          
    
    slices = {}   
    
    head = []
    abdome = []
    
    air_h = 0
    water_h = 0
    air_b = 0
    water_b = 0        
   
    # Print the files    
    for elem in listOfFiles:
        #print(elem) 
        try:
            #print("DICOM")
            ds = pydicom.dcmread(elem)
            
            #DICOM DERIVADO do ORIGINAL
            ds.ImageType[0] = 'DERIVED'
            
            #URL
            ds.add_new(0x00081190, 'UR', elem)
             
            
            #print(ds.ProtocolName)
            if (np.mean((crop_center(ds.pixel_array,100,100))) < -500 and air_h == 0 and 'CRANIO' in ds.ProtocolName):
                #print('ar')
                ds.SeriesDescription = "Air" 
                head.append(ds)            
                #print("----------------------")
                air_h = 1
            elif(water_h == 0 and 'CRANIO' in ds.ProtocolName):
                #print('agua')
                ds.SeriesDescription = "Walter" 
                head.append(ds)
                water_h = 1
            elif (np.mean((crop_center(ds.pixel_array,100,100))) < -500 and air_b == 0 and 'ABDOME' in ds.ProtocolName):
                #print('ar')
                ds.SeriesDescription = "Air" 
                abdome.append(ds)
                #print(ds.ProtocolName)
                #print("----------------------")
                air_b = 1
            elif(water_b == 0 and 'ABDOME' in ds.ProtocolName):
                #print('agua')
                ds.SeriesDescription = "Walter" 
                abdome.append(ds)
                water_b = 1                                     
        except:
            logging.warning("Não é DICOM")
            
    slices = {"CRANIO": head, "ABDOME": abdome}
    #pprint.pprint(slices)
    
    #print('----------------')
 
    '''         
    slices.sort(key = lambda x: int(x.InstanceNumber))
    try:
        slice_thickness = np.abs(slices[0].ImagePositionPatient[2] - slices[1].ImagePositionPatient[2])
    except:
        slice_thickness = np.abs(slices[0].SliceLocation - slices[1].SliceLocation)
        
    for s in slices:
        s.SliceThickness = slice_thickness 
    '''
        
    return slices

def get_pixels_hu(s):
    image = np.stack(s.pixel_array)
    # Convert to int16 (from sometimes int16), 
    # should be possible as values should always be low enough (<32k)
    image = image.astype(np.int16)

    # Set outside-of-scan pixels to 1
    # The intercept is usually -1024, so air is approximately 0
    image[image == -2000] = 0
    
    # Convert to Hounsfield units (HU)
    intercept = s.RescaleIntercept
    slope = s.RescaleSlope
    
    if slope != 1:
        image = slope * image.astype(np.float64)
        image = image.astype(np.int16)
        
    image += np.int16(intercept)
    
    return np.array(image, dtype=np.int16)

def get_pixels_hu_scans(scans):
    image = np.stack([s.pixel_array for s in scans])
    # Convert to int16 (from sometimes int16), 
    # should be possible as values should always be low enough (<32k)
    image = image.astype(np.int16)

    # Set outside-of-scan pixels to 1
    # The intercept is usually -1024, so air is approximately 0
    image[image == -2000] = 0
    
    # Convert to Hounsfield units (HU)
    intercept = scans[0].RescaleIntercept
    slope = scans[0].RescaleSlope
    
    if slope != 1:
        image = slope * image.astype(np.float64)
        image = image.astype(np.int16)
        
    image += np.int16(intercept)
    
    return np.array(image, dtype=np.int16)


def crop_center(img,cropx,cropy):
    #https://stackoverflow.com/questions/39382412/crop-center-portion-of-a-numpy-image
    y,x = img.shape
 
    startx = x//2-(cropx//2)
    starty = y//2-(cropy//2)  
    return img[starty:starty+cropy,startx:startx+cropx]

def crop_sup_left(img,cropx,cropy):
    y,x = img.shape
    startx = x//3-(cropx//2)
    starty = y//3-(cropy//2)
    return img[starty:starty+cropy,startx:startx+cropx]

def crop_inf_left(img,cropx,cropy):
    y,x = img.shape
    startx = (x//2)//3-(cropx//2)
    starty = (y//3)-(cropy//2)    
    return img[starty:starty+cropy,startx:startx+cropx]


def crop_sup_rig(img,cropx,cropy):
    y,x = img.shape
    startx = (x//3)-(cropx//2)
    starty = (y//2)//3-(cropy//2)    
    return img[starty:starty+cropy,startx:startx+cropx]

def crop_inf_rig(img,cropx,cropy):
    y,x = img.shape
    startx = (x//2)//3-(cropx//2)
    starty = (y//2)//3-(cropy//2)    
    return img[starty:starty+cropy,startx:startx+cropx]

def resuRuido(Desvio):
    re = ""
    if(Desvio>20):
        re ="Nível de suspensão"
    elif(Desvio>15):
        re ="Não conforme"
    else:
        re ="Conforme"
    return re


def resuUniformidade(Desvio):
    resuUniformidade = ""
    if(Desvio <= 5):
        resuUniformidade = "Conforme"
    elif(Desvio > 10):
        resuUniformidade = "Nível de Suspensão" 
    else:
        resuUniformidade = "Não Conforme"
    return resuUniformidade

def resuMeanCT(ds,val):
    re = ""
    if(ds == "Air" and abs(val) <= 1004):
        re = "Conforme"
    elif(ds == "Walter" and abs(val) <=4 ):
        re = "Conforme"
    else:
        re = "Não Conforme"
        
    return re
       


def get_rois(ds,roi,locale):
    values ={}
        
    values ={
            "mean" :np.mean((locale(ds,roi,roi))),
            "std" : np.std((locale(ds,roi,roi))),
            "min" : np.min((locale(ds,roi,roi))),
            "max" : np.max((locale(ds,roi,roi))),
            "noise": (np.std((crop_center(ds,roi,roi)))*100)/1000,
            "uniformity": np.mean((crop_center(ds,roi,roi)))- np.mean((locale(ds,roi,roi))) ,
    }    
    return values
    
def plot(imgs):
    print("file count: {}".format(len(imgs)))
    print(imgs)
    for ds in imgs:        
        roi = 100
        
        fileDada = {
            "crop_center":   get_rois(ds,roi,crop_center),
            "crop_sup_left": get_rois(ds,roi,crop_sup_left) ,
            "crop_inf_left": get_rois(ds,roi,crop_inf_left),
            "crop_sup_rig":  get_rois(ds,roi,crop_sup_rig),
            "crop_inf_rig":  get_rois(ds,roi,crop_sup_rig), 
        }
        
        pprint.pprint(fileDada)

        
 
              
        plt.imshow(crop_center(ds,roi,roi), cmap=pylab.cm.bone) # shows image with mask
        plt.show()  


    
def get_roi_data(ds):
      
    roi = 100
        
    fileDada = {
            "center":   get_rois(ds,roi,crop_center),
            "sup_left": get_rois(ds,roi,crop_sup_left) ,
            "inf_left": get_rois(ds,roi,crop_inf_left) ,
            "sup_rig":  get_rois(ds,roi,crop_sup_rig) ,
            "inf_rig":  get_rois(ds,roi,crop_sup_rig) 
    }
    
    maxNoise = []
    maxUniformity = []
    maxMean = []
    for key, val in fileDada.items():
        maxUniformity.append(val["uniformity"])
        maxNoise.append(val["noise"])
        maxMean.append(val["mean"])
    #pprint.pprint(max(list(maxNoise)))
    
    #https://www.seram.es/images/site/protocolo_2011.pdf
    #TC014.- Valor medio del número CT
    fileDada["max_mean"] = max(list(maxMean))

    fileDada["max_noise"] = max(list(maxNoise))
    fileDada["max_uniformity"] = max(list(maxUniformity))
    fileDada["resu_noise"] = resuRuido(max(list(maxNoise)))
    fileDada["resu_uniformity"] = resuUniformidade(max(list(maxUniformity)))
    
    #print(ds)
    #fileDada["resu_mean"] = resuMeanCT(ds,max(list(maxMean)))
    
    #get max value from Noise and Uniformity
    #https://stackoverflow.com/questions/13067615/getting-the-max-value-of-attributes-from-a-list-of-objects
    #print(max(node.center for node in fileDada))
  
    return fileDada

        
        


def get_array_measure(slices):
    protocol = {}
    for key, val in slices.items():
        
        pr = {}
        for a in val:
            #print(a)
            slicedata ={}
            hu = get_pixels_hu(a)
            rois = get_roi_data(hu)

            slicedata = {
                'SeriesDescription':a.SeriesDescription,
                'StudyDate': a.StudyDate,
                'SeriesDate': a.SeriesDate,
                'AcquisitionDate':a.AcquisitionDate,
                'AcquisitionDate':a.AcquisitionDate,  
                'BodyPartExamined':a.BodyPartExamined, 
                'ScanOptions':a.ScanOptions,  
                'SliceThickness':a.SliceThickness,
                'KVP':a.KVP,
                'mA': a[0x0018, 0x1151].value,
                'DataCollectionDiameter':a.DataCollectionDiameter,  
                'ProtocolName':a.ProtocolName, 
                'ReconstructionDiameter':a.ReconstructionDiameter,
                'ExposureTime':a.ExposureTime,
                'Exposure':a.Exposure,
                'CTDIvol':a.CTDIvol,
                'RevolutionTime':a.RevolutionTime,
                'TotalCollimationWidth':a.TotalCollimationWidth,
                'Manufacturer':a.Manufacturer,
                'Model': a[0x0008, 0x1090].value, 
                'StationName':a.StationName,
                'DeviceSerialNumber':a.DeviceSerialNumber,
                'InstitutionName':a.InstitutionName,
                'InstitutionAddress':a.InstitutionAddress,
                'RetrieveURL':a[0x0008, 0x1190].value,
                
            
                


                'rois':rois
            }


            pr[a.SeriesDescription] = slicedata
            protocol[key] = pr
            #print(slicedata)
    return protocol

def processCt(folder):
		slices = load_scan(folder)
		#imgs = get_pixels_hu_scans(slices["ABDOME"])
		dados = get_array_measure(slices)
		#pprint.pprint(dados)


		roi_data = dados
		rn =[]
		ru =[]
		rc =[]
		for key, value in dados.items():
		    for ke, val in value.items():
		        ru.append(val["rois"]["resu_uniformity"])
		        rn.append(val["rois"]["resu_noise"])
		        rc.append(resuMeanCT(val["SeriesDescription"],val["rois"]["max_mean"]))

		roi_data["uniformity"] = "Não Conforme" if 'Não Conforme' in ru else "Conforme"
		roi_data["noise"] = "Não Conforme" if 'Não Conforme' in rn else "Conforme"
		roi_data["ct_number"] = "Não Conforme" if 'Não Conforme' in rc else "Conforme"
		#print('--------')
		#print(ru)
		#print(rn)
		return roi_data

		#plot(imgs)
		#print(len(slices)) 
class ProcessingCt:
        def __init__(self):
            self.file_path = None
            self.fantoma = None
            self.nome_protocolo_1=None
            self.tensao_1=None
            self.corrente_1=None
            self.rotation_time_1=None
            self.pitch_1=None
            self.reconstrucao_1=None
            self.incremento_1=None
            self.espessura_corte_1=None
            self.nome_protocolo_2=None
            self.tensao_2=None
            self.corrente_2=None
            self.rotation_time_2=None
            self.pitch_2=None
            self.reconstrucao_2=None
            self.incremento_2=None
            self.espessura_corte_2=None
            self.wadoURL1=None
            self.wadoURL2=None
            self.wadoURL3=None
            self.wadoURL4=None
            self.wadoURL5=None
            self.wadoURL6=None
            self.jsonconfig=None
            self.area_cr_a_1=None
            self.area_cr_a_2=None
            self.area_cr_a_3=None
            self.area_cr_a_4=None
            self.area_cr_a_5=None
            self.media_cr_a_1=None
            self.media_cr_a_2=None
            self.media_cr_a_3=None
            self.media_cr_a_4=None
            self.media_cr_a_5=None
            self.dspad_cr_a_1=None
            self.dspad_cr_a_2=None
            self.dspad_cr_a_3=None
            self.dspad_cr_a_4=None
            self.dspad_cr_a_5=None
            self.unif_cr_a=None
            self.res_unif_cr_a=None
            self.ruido_cr_a=None
            self.res_ruido_cr_a=None
            self.area_ab_a_1=None
            self.area_ab_a_2=None
            self.area_ab_a_3=None
            self.area_ab_a_4=None
            self.area_ab_a_5=None
            self.media_ab_a_1=None
            self.media_ab_a_2=None
            self.media_ab_a_3=None
            self.media_ab_a_4=None
            self.media_ab_a_5=None
            self.dspad_ab_a_1=None
            self.dspad_ab_a_2=None
            self.dspad_ab_a_3=None
            self.dspad_ab_a_4=None
            self.dspad_ab_a_5=None
            self.unif_ab_a=None
            self.res_unif_ab_a=None
            self.ruido_ab_a=None
            self.res_ruido_ab_a=None
            self.area_cr_ar_1=None
            self.area_cr_ar_2=None
            self.area_cr_ar_3=None
            self.area_cr_ar_4=None
            self.area_cr_ar_5=None
            self.media_cr_ar_1=None
            self.media_cr_ar_2=None
            self.media_cr_ar_3=None
            self.media_cr_ar_4=None
            self.media_cr_ar_5=None
            self.dspad_cr_ar_1=None
            self.dspad_cr_ar_2=None
            self.dspad_cr_ar_3=None
            self.dspad_cr_ar_4=None
            self.dspad_cr_ar_5=None
            self.unif_cr_ar=None
            self.res_unif_cr_ar=None
            self.ruido_cr_ar=None
            self.res_ruido_cr_ar=None
            self.area_ab_ar_1=None
            self.area_ab_ar_2=None
            self.area_ab_ar_3=None
            self.area_ab_ar_4=None
            self.area_ab_ar_5=None
            self.media_ab_ar_1=None
            self.media_ab_ar_2=None
            self.media_ab_ar_3=None
            self.media_ab_ar_4=None
            self.media_ab_ar_5=None
            self.dspad_ab_ar_1=None
            self.dspad_ab_ar_2=None
            self.dspad_ab_ar_3=None
            self.dspad_ab_ar_4=None
            self.dspad_ab_ar_5=None
            self.unif_ab_ar=None
            self.res_unif_ab_ar=None
            self.ruido_ab_ar=None
            self.res_ruido_ab_ar=None
            self.nct_med_cr_1=None
            self.nct_med_cr_2=None
            self.nct_med_cr_3=None
            self.nct_med_cr_4=None
            self.nct_med_cr_5=None
            self.nct_dspad_cr_1=None
            self.nct_dspad_cr_2=None
            self.nct_dspad_cr_3=None
            self.nct_dspad_cr_4=None
            self.nct_dspad_cr_5=None
            self.res_ar_cr_sc=None
            self.res_agua_cr_sc=None
            self.res_osso_cr=None
            self.res_ar_cr=None
            self.res_agua_cr=None
            self.res_acrilico_cr=None
            self.res_polietileno_cr=None
            self.nct_med_ab_1=None
            self.nct_med_ab_2=None
            self.nct_med_ab_3=None
            self.nct_med_ab_4=None
            self.nct_med_ab_5=None
            self.nct_dspad_ab_1=None
            self.nct_dspad_ab_2=None
            self.nct_dspad_ab_3=None
            self.nct_dspad_ab_4=None
            self.nct_dspad_ab_5=None
            self.res_ar_ab_sc=None
            self.res_agua_ab_sc=None
            self.res_osso_ab=None
            self.res_ar_ab=None
            self.res_agua_ab=None
            self.res_acrilico_ab=None
            self.res_polietileno_ab=None
            self.obs=None
            self.nct_dspad_ab_6=None
            self.nct_med_ab_6=None
            self.nct_dspad_cr_6=None
            self.nct_med_cr_6=None
            self.nct_med_rel_ab_1=None
            self.nct_med_rel_ab_2=None
            self.nct_med_rel_ab_3=None
            self.nct_med_rel_ab_4=None
            self.nct_med_rel_ab_5=None
            self.nct_med_rel_ab_6=None
            self.nct_med_rel_cr_1=None
            self.nct_med_rel_cr_2=None
            self.nct_med_rel_cr_3=None
            self.nct_med_rel_cr_4=None
            self.nct_med_rel_cr_5=None
            self.nct_med_rel_cr_6=None
            self.resultado=None
            self.cliente_id=None
            self.equipamento_id=None
            self.user_id=None
            self.visualized=None
            self.print=None
            self.print_date=None
            self.print_user=None
            self.relatorio=None
            self.unf_cr_a_1=None
            self.unf_cr_a_2=None
            self.unf_cr_a_3=None
            self.unf_cr_a_4=None
            self.unf_cr_a_5=None
            self.ruido_cr_a_1=None
            self.ruido_cr_a_2=None
            self.ruido_cr_a_3=None
            self.ruido_cr_a_4=None
            self.ruido_cr_a_5=None
            self.unf_ab_a_1=None
            self.unf_ab_a_2=None
            self.unf_ab_a_3=None
            self.unf_ab_a_4=None
            self.unf_ab_a_5=None
            self.ruido_ab_a_1=None
            self.ruido_ab_a_2=None
            self.ruido_ab_a_3=None
            self.ruido_ab_a_4=None
            self.ruido_ab_a_5=None
            self.unf_cr_ar_1=None
            self.unf_cr_ar_2=None
            self.unf_cr_ar_3=None
            self.unf_cr_ar_4=None
            self.unf_cr_ar_5=None
            self.ruido_cr_ar_1=None
            self.ruido_cr_ar_2=None
            self.ruido_cr_ar_3=None
            self.ruido_cr_ar_4=None
            self.ruido_cr_ar_5=None
            self.unf_ab_ar_1=None
            self.unf_ab_ar_2=None
            self.unf_ab_ar_3=None
            self.unf_ab_ar_4=None
            self.unf_ab_ar_5=None
            self.ruido_ab_ar_1=None
            self.ruido_ab_ar_2=None
            self.ruido_ab_ar_3=None
            self.ruido_ab_ar_4=None
            self.ruido_ab_ar_5=None

        def contructObj(self,obj):
         
        
            self.wadoURL1= obj['CRANIO']['Walter']['RetrieveURL'] 
            self.wadoURL2= obj['ABDOME']['Walter']['RetrieveURL'] 
            self.wadoURL3= obj['CRANIO']['Air']['RetrieveURL'] 
            self.wadoURL4= obj['ABDOME']['Air']['RetrieveURL'] 
            #self.wadoURL5= obj['CRANIO']['Walter']['RetrieveURL'] 
            #self.wadoURL6= obj['CRANIO']['Walter']['RetrieveURL']         
        
            #Nome Protocolo (Crânio)
            
            self.nome_protocolo_1=obj['CRANIO']['Walter']['ProtocolName']
            self.tensao_1=obj['CRANIO']['Walter']['KVP']
            self.corrente_1=obj['CRANIO']['Walter']['mA']
            self.rotation_time_1=obj['CRANIO']['Walter']['RevolutionTime']
            self.pitch_1=None
            self.reconstrucao_1=None
            self.incremento_1=None
            self.espessura_corte_1=obj['CRANIO']['Walter']['SliceThickness']
            
            # Crânio Corte na Água
            #AREA 
            self.area_cr_a_1=100
            self.area_cr_a_2=100
            self.area_cr_a_3=100
            self.area_cr_a_4=100
            self.area_cr_a_5=100
            #MEDIA
            self.media_cr_a_1=obj['CRANIO']['Walter']['rois']['sup_left']['mean']     #sup_left
            self.media_cr_a_2=obj['CRANIO']['Walter']['rois']['sup_rig']['mean']     #sup_rig
            self.media_cr_a_3=obj['CRANIO']['Walter']['rois']['center']['mean']     #center
            self.media_cr_a_4=obj['CRANIO']['Walter']['rois']['inf_rig']['mean']     #inf_rig
            self.media_cr_a_5=obj['CRANIO']['Walter']['rois']['inf_left']['mean']     #inf_left
            #DESVIO
            self.dspad_cr_a_1=obj['CRANIO']['Walter']['rois']['sup_left']['std']     #sup_left
            self.dspad_cr_a_2=obj['CRANIO']['Walter']['rois']['sup_rig']['std']     #sup_rig
            self.dspad_cr_a_3=obj['CRANIO']['Walter']['rois']['center']['std']     #center
            self.dspad_cr_a_4=obj['CRANIO']['Walter']['rois']['inf_rig']['std']     #inf_rig
            self.dspad_cr_a_5=obj['CRANIO']['Walter']['rois']['inf_left']['std']      #inf_left
            #UNIFORMIDADE
            self.unf_cr_a_1=obj['CRANIO']['Walter']['rois']['sup_left']['uniformity']        #sup_left
            self.unf_cr_a_2=obj['CRANIO']['Walter']['rois']['sup_rig']['uniformity']        #sup_rig
            self.unf_cr_a_3=obj['CRANIO']['Walter']['rois']['center']['uniformity']        #center
            self.unf_cr_a_4=obj['CRANIO']['Walter']['rois']['inf_rig']['uniformity']        #inf_rig
            self.unf_cr_a_5=obj['CRANIO']['Walter']['rois']['inf_left']['uniformity']        #inf_left
            #RUIDO
            self.ruido_cr_a_1=obj['CRANIO']['Walter']['rois']['sup_left']['uniformity']      #sup_left
            self.ruido_cr_a_2=obj['CRANIO']['Walter']['rois']['sup_rig']['uniformity']      #sup_rig
            self.ruido_cr_a_3=obj['CRANIO']['Walter']['rois']['center']['uniformity']      #center
            self.ruido_cr_a_4=obj['CRANIO']['Walter']['rois']['inf_rig']['uniformity']      #inf_rig
            self.ruido_cr_a_5=obj['CRANIO']['Walter']['rois']['inf_left']['uniformity']      #inf_left
            #RESULTADOS:
            #uniformidade
            self.unif_cr_a=obj['CRANIO']['Walter']['rois']['max_mean']
            #res uniformidade
            self.res_unif_cr_a=obj['CRANIO']['Walter']['rois']['resu_uniformity']

            #ruido
            self.ruido_cr_a=obj['CRANIO']['Walter']['rois']['max_noise']
            #res ruido
            self.res_ruido_cr_a=obj['CRANIO']['Walter']['rois']['resu_noise']
            

            #- Crânio Corte na Ar
            #AREA 
            self.area_cr_ar_1=100
            self.area_cr_ar_2=100
            self.area_cr_ar_3=100
            self.area_cr_ar_4=100
            self.area_cr_ar_5=100
            #MEDIA
            self.media_cr_ar_1=obj['CRANIO']['Air']['rois']['sup_left']['mean']     #sup_left
            self.media_cr_ar_2=obj['CRANIO']['Air']['rois']['sup_rig']['mean']     #sup_rig
            self.media_cr_ar_3=obj['CRANIO']['Air']['rois']['center']['mean']     #center
            self.media_cr_ar_4=obj['CRANIO']['Air']['rois']['inf_rig']['mean']     #inf_rig
            self.media_cr_ar_5=obj['CRANIO']['Air']['rois']['inf_left']['mean']     #inf_left
            #DESVIO
            self.dspad_cr_ar_1=obj['CRANIO']['Air']['rois']['sup_left']['std']     #sup_left
            self.dspad_cr_ar_2=obj['CRANIO']['Air']['rois']['sup_rig']['std']     #sup_rig
            self.dspad_cr_ar_3=obj['CRANIO']['Air']['rois']['center']['std']     #center
            self.dspad_cr_ar_4=obj['CRANIO']['Air']['rois']['inf_rig']['std']     #inf_rig
            self.dspad_cr_ar_5=obj['CRANIO']['Air']['rois']['inf_left']['std']      #inf_left
            #UNIFORMIDADE
            self.unf_cr_ar_1=obj['CRANIO']['Air']['rois']['sup_left']['uniformity']        #sup_left
            self.unf_cr_ar_2=obj['CRANIO']['Air']['rois']['sup_rig']['uniformity']        #sup_rig
            self.unf_cr_ar_3=obj['CRANIO']['Air']['rois']['center']['uniformity']        #center
            self.unf_cr_ar_4=obj['CRANIO']['Air']['rois']['inf_rig']['uniformity']        #inf_rig
            self.unf_cr_ar_5=obj['CRANIO']['Air']['rois']['inf_left']['uniformity']        #inf_left
            #RUIDO
            self.ruido_cr_ar_1=obj['CRANIO']['Air']['rois']['sup_left']['uniformity']      #sup_left
            self.ruido_cr_ar_2=obj['CRANIO']['Air']['rois']['sup_rig']['uniformity']      #sup_rig
            self.ruido_cr_ar_3=obj['CRANIO']['Air']['rois']['center']['uniformity']      #center
            self.ruido_cr_ar_4=obj['CRANIO']['Air']['rois']['inf_rig']['uniformity']      #inf_rig
            self.ruido_cr_ar_5=obj['CRANIO']['Air']['rois']['inf_left']['uniformity']      #inf_left
            #RESULTADOS:
            #uniformidade
            self.unif_cr_ar=obj['CRANIO']['Air']['rois']['max_mean']
            #res uniformidade
            self.res_unif_cr_ar=obj['CRANIO']['Air']['rois']['resu_uniformity']

            #ruido
            self.ruido_cr_ar=obj['CRANIO']['Air']['rois']['max_noise']
            #res ruido
            self.res_ruido_cr_ar=obj['CRANIO']['Air']['rois']['resu_noise']
                
            


            #Nome Protocolo (Abdome)
            self.nome_protocolo_2=obj['ABDOME']['Walter']['ProtocolName']
            self.tensao_2=obj['ABDOME']['Walter']['KVP']
            self.corrente_2=obj['ABDOME']['Walter']['mA']
            self.rotation_time_2=obj['ABDOME']['Walter']['RevolutionTime']
            self.pitch_2=None
            self.reconstrucao_2=None
            self.incremento_2=None
            self.espessura_corte_2=obj['ABDOME']['Walter']['SliceThickness']
            
            
            #Abdome Corte na Água
            #AREA 
            self.area_ab_a_1=100
            self.area_ab_a_2=100
            self.area_ab_a_3=100
            self.area_ab_a_4=100
            self.area_ab_a_5=100
            #MEDIA
            self.media_ab_a_1=obj['ABDOME']['Walter']['rois']['sup_left']['mean']     #sup_left
            self.media_ab_a_2=obj['ABDOME']['Walter']['rois']['sup_rig']['mean']     #sup_rig
            self.media_ab_a_3=obj['ABDOME']['Walter']['rois']['center']['mean']     #center
            self.media_ab_a_4=obj['ABDOME']['Walter']['rois']['inf_rig']['mean']     #inf_rig
            self.media_ab_a_5=obj['ABDOME']['Walter']['rois']['inf_left']['mean']     #inf_left
            #DESVIO
            self.dspad_ab_a_1=obj['ABDOME']['Walter']['rois']['sup_left']['std']     #sup_left
            self.dspad_ab_a_2=obj['ABDOME']['Walter']['rois']['sup_rig']['std']     #sup_rig
            self.dspad_ab_a_3=obj['ABDOME']['Walter']['rois']['center']['std']     #center
            self.dspad_ab_a_4=obj['ABDOME']['Walter']['rois']['inf_rig']['std']     #inf_rig
            self.dspad_ab_a_5=obj['ABDOME']['Walter']['rois']['inf_left']['std']      #inf_left
            #UNIFORMIDADE
            self.unf_ab_a_1=obj['ABDOME']['Walter']['rois']['sup_left']['uniformity']        #sup_left
            self.unf_ab_a_2=obj['ABDOME']['Walter']['rois']['sup_rig']['uniformity']        #sup_rig
            self.unf_ab_a_3=obj['ABDOME']['Walter']['rois']['center']['uniformity']        #center
            self.unf_ab_a_4=obj['ABDOME']['Walter']['rois']['inf_rig']['uniformity']        #inf_rig
            self.unf_ab_a_5=obj['ABDOME']['Walter']['rois']['inf_left']['uniformity']        #inf_left
            #RUIDO
            self.ruido_ab_a_1=obj['ABDOME']['Walter']['rois']['sup_left']['uniformity']      #sup_left
            self.ruido_ab_a_2=obj['ABDOME']['Walter']['rois']['sup_rig']['uniformity']      #sup_rig
            self.ruido_ab_a_3=obj['ABDOME']['Walter']['rois']['center']['uniformity']      #center
            self.ruido_ab_a_4=obj['ABDOME']['Walter']['rois']['inf_rig']['uniformity']      #inf_rig
            self.ruido_ab_a_5=obj['ABDOME']['Walter']['rois']['inf_left']['uniformity']      #inf_left
            #RESULTADOS:
            
            #uniformidade
            self.unif_ab_a=obj['ABDOME']['Walter']['rois']['max_mean']
            #res uniformidade
            self.res_unif_ab_a=obj['ABDOME']['Walter']['rois']['resu_uniformity']

            #ruido
            self.ruido_ab_a=obj['ABDOME']['Walter']['rois']['max_noise']
            #res ruido
            self.res_ruido_ab_a=obj['ABDOME']['Walter']['rois']['resu_noise']
            

            #Abdome Corte na Ar
            #AREA 
            self.area_ab_ar_1=100
            self.area_ab_ar_2=100
            self.area_ab_ar_3=100
            self.area_ab_ar_4=100
            self.area_ab_ar_5=100
            #MEDIA
            self.media_ab_ar_1=obj['ABDOME']['Air']['rois']['sup_left']['mean']     #sup_left
            self.media_ab_ar_2=obj['ABDOME']['Air']['rois']['sup_rig']['mean']     #sup_rig
            self.media_ab_ar_3=obj['ABDOME']['Air']['rois']['center']['mean']     #center
            self.media_ab_ar_4=obj['ABDOME']['Air']['rois']['inf_rig']['mean']     #inf_rig
            self.media_ab_ar_5=obj['ABDOME']['Air']['rois']['inf_left']['mean']     #inf_left
            #DESVIO
            self.dspad_ab_ar_1=obj['ABDOME']['Air']['rois']['sup_left']['std']     #sup_left
            self.dspad_ab_ar_2=obj['ABDOME']['Air']['rois']['sup_rig']['std']     #sup_rig
            self.dspad_ab_ar_3=obj['ABDOME']['Air']['rois']['center']['std']     #center
            self.dspad_ab_ar_4=obj['ABDOME']['Air']['rois']['inf_rig']['std']     #inf_rig
            self.dspad_ab_ar_5=obj['ABDOME']['Air']['rois']['inf_left']['std']      #inf_left
            #UNIFORMIDADE
            self.unf_ab_ar_1=obj['ABDOME']['Air']['rois']['sup_left']['uniformity']        #sup_left
            self.unf_ab_ar_2=obj['ABDOME']['Air']['rois']['sup_rig']['uniformity']        #sup_rig
            self.unf_ab_ar_3=obj['ABDOME']['Air']['rois']['center']['uniformity']        #center
            self.unf_ab_ar_4=obj['ABDOME']['Air']['rois']['inf_rig']['uniformity']        #inf_rig
            self.unf_ab_ar_5=obj['ABDOME']['Air']['rois']['inf_left']['uniformity']        #inf_left
            #RUIDO
            self.ruido_ab_ar_1=obj['ABDOME']['Air']['rois']['sup_left']['uniformity']      #sup_left
            self.ruido_ab_ar_2=obj['ABDOME']['Air']['rois']['sup_rig']['uniformity']      #sup_rig
            self.ruido_ab_ar_3=obj['ABDOME']['Air']['rois']['center']['uniformity']      #center
            self.ruido_ab_ar_4=obj['ABDOME']['Air']['rois']['inf_rig']['uniformity']      #inf_rig
            self.ruido_ab_ar_5=obj['ABDOME']['Air']['rois']['inf_left']['uniformity']      #inf_left
            #RESULTADOS:
            
            #uniformidade
            self.unif_ab_ar=obj['ABDOME']['Air']['rois']['max_mean']
            #res uniformidade
            self.res_unif_ab_ar=obj['ABDOME']['Air']['rois']['resu_uniformity']

            #ruido
            self.ruido_ab_ar=obj['ABDOME']['Air']['rois']['max_noise']
            #res ruido
            self.res_ruido_ab_ar=obj['ABDOME']['Air']['rois']['resu_noise']
            

            
        def calculate(self):
 
            path = self.file_path
            re = processCt(path)
            self.contructObj(re)


            resultado = {
                        'nome_protocolo_1':self.nome_protocolo_1,
                        'tensao_1':self.tensao_1,
                        'corrente_1':self.corrente_1,
                        'rotation_time_1':self.rotation_time_1,
                        'pitch_1':self.pitch_1,
                        'reconstrucao_1':self.reconstrucao_1,
                        'incremento_1':self.incremento_1,
                        'espessura_corte_1':self.espessura_corte_1,
                        'nome_protocolo_2':self.nome_protocolo_2,
                        'tensao_2':self.tensao_2,
                        'corrente_2':self.corrente_2,
                        'rotation_time_2':self.rotation_time_2,
                        'pitch_2':self.pitch_2,
                        'reconstrucao_2':self.reconstrucao_2,
                        'incremento_2':self.incremento_2,
                        'espessura_corte_2':self.espessura_corte_2,
                        'wadoURL1':self.wadoURL1,
                        'wadoURL2':self.wadoURL2,
                        'wadoURL3':self.wadoURL3,
                        'wadoURL4':self.wadoURL4,
                        'wadoURL5':self.wadoURL5,
                        'wadoURL6':self.wadoURL6,
                        'jsonconfig':self.jsonconfig,
                        'area_cr_a_1':self.area_cr_a_1,
                        'area_cr_a_2':self.area_cr_a_2,
                        'area_cr_a_3':self.area_cr_a_3,
                        'area_cr_a_4':self.area_cr_a_4,
                        'area_cr_a_5':self.area_cr_a_5,
                        'media_cr_a_1':self.media_cr_a_1,
                        'media_cr_a_2':self.media_cr_a_2,
                        'media_cr_a_3':self.media_cr_a_3,
                        'media_cr_a_4':self.media_cr_a_4,
                        'media_cr_a_5':self.media_cr_a_5,
                        'dspad_cr_a_1':self.dspad_cr_a_1,
                        'dspad_cr_a_2':self.dspad_cr_a_2,
                        'dspad_cr_a_3':self.dspad_cr_a_3,
                        'dspad_cr_a_4':self.dspad_cr_a_4,
                        'dspad_cr_a_5':self.dspad_cr_a_5,
                        'unif_cr_a':self.unif_cr_a,
                        'res_unif_cr_a':self.res_unif_cr_a,
                        'ruido_cr_a':self.ruido_cr_a,
                        'res_ruido_cr_a':self.res_ruido_cr_a,
                        'area_ab_a_1':self.area_ab_a_1,
                        'area_ab_a_2':self.area_ab_a_2,
                        'area_ab_a_3':self.area_ab_a_3,
                        'area_ab_a_4':self.area_ab_a_4,
                        'area_ab_a_5':self.area_ab_a_5,
                        'media_ab_a_1':self.media_ab_a_1,
                        'media_ab_a_2':self.media_ab_a_2,
                        'media_ab_a_3':self.media_ab_a_3,
                        'media_ab_a_4':self.media_ab_a_4,
                        'media_ab_a_5':self.media_ab_a_5,
                        'dspad_ab_a_1':self.dspad_ab_a_1,
                        'dspad_ab_a_2':self.dspad_ab_a_2,
                        'dspad_ab_a_3':self.dspad_ab_a_3,
                        'dspad_ab_a_4':self.dspad_ab_a_4,
                        'dspad_ab_a_5':self.dspad_ab_a_5,
                        'unif_ab_a':self.unif_ab_a,
                        'res_unif_ab_a':self.res_unif_ab_a,
                        'ruido_ab_a':self.ruido_ab_a,
                        'res_ruido_ab_a':self.res_ruido_ab_a,
                        'area_cr_ar_1':self.area_cr_ar_1,
                        'area_cr_ar_2':self.area_cr_ar_2,
                        'area_cr_ar_3':self.area_cr_ar_3,
                        'area_cr_ar_4':self.area_cr_ar_4,
                        'area_cr_ar_5':self.area_cr_ar_5,
                        'media_cr_ar_1':self.media_cr_ar_1,
                        'media_cr_ar_2':self.media_cr_ar_2,
                        'media_cr_ar_3':self.media_cr_ar_3,
                        'media_cr_ar_4':self.media_cr_ar_4,
                        'media_cr_ar_5':self.media_cr_ar_5,
                        'dspad_cr_ar_1':self.dspad_cr_ar_1,
                        'dspad_cr_ar_2':self.dspad_cr_ar_2,
                        'dspad_cr_ar_3':self.dspad_cr_ar_3,
                        'dspad_cr_ar_4':self.dspad_cr_ar_4,
                        'dspad_cr_ar_5':self.dspad_cr_ar_5,
                        'unif_cr_ar':self.unif_cr_ar,
                        'res_unif_cr_ar':self.res_unif_cr_ar,
                        'ruido_cr_ar':self.ruido_cr_ar,
                        'res_ruido_cr_ar':self.res_ruido_cr_ar,
                        'area_ab_ar_1':self.area_ab_ar_1,
                        'area_ab_ar_2':self.area_ab_ar_2,
                        'area_ab_ar_3':self.area_ab_ar_3,
                        'area_ab_ar_4':self.area_ab_ar_4,
                        'area_ab_ar_5':self.area_ab_ar_5,
                        'media_ab_ar_1':self.media_ab_ar_1,
                        'media_ab_ar_2':self.media_ab_ar_2,
                        'media_ab_ar_3':self.media_ab_ar_3,
                        'media_ab_ar_4':self.media_ab_ar_4,
                        'media_ab_ar_5':self.media_ab_ar_5,
                        'dspad_ab_ar_1':self.dspad_ab_ar_1,
                        'dspad_ab_ar_2':self.dspad_ab_ar_2,
                        'dspad_ab_ar_3':self.dspad_ab_ar_3,
                        'dspad_ab_ar_4':self.dspad_ab_ar_4,
                        'dspad_ab_ar_5':self.dspad_ab_ar_5,
                        'unif_ab_ar':self.unif_ab_ar,
                        'res_unif_ab_ar':self.res_unif_ab_ar,
                        'ruido_ab_ar':self.ruido_ab_ar,
                        'res_ruido_ab_ar':self.res_ruido_ab_ar,
                        'nct_med_cr_1':self.nct_med_cr_1,
                        'nct_med_cr_2':self.nct_med_cr_2,
                        'nct_med_cr_3':self.nct_med_cr_3,
                        'nct_med_cr_4':self.nct_med_cr_4,
                        'nct_med_cr_5':self.nct_med_cr_5,
                        'nct_dspad_cr_1':self.nct_dspad_cr_1,
                        'nct_dspad_cr_2':self.nct_dspad_cr_2,
                        'nct_dspad_cr_3':self.nct_dspad_cr_3,
                        'nct_dspad_cr_4':self.nct_dspad_cr_4,
                        'nct_dspad_cr_5':self.nct_dspad_cr_5,
                        'res_ar_cr_sc':self.res_ar_cr_sc,
                        'res_agua_cr_sc':self.res_agua_cr_sc,
                        'res_osso_cr':self.res_osso_cr,
                        'res_ar_cr':self.res_ar_cr,
                        'res_agua_cr':self.res_agua_cr,
                        'res_acrilico_cr':self.res_acrilico_cr,
                        'res_polietileno_cr':self.res_polietileno_cr,
                        'nct_med_ab_1':self.nct_med_ab_1,
                        'nct_med_ab_2':self.nct_med_ab_2,
                        'nct_med_ab_3':self.nct_med_ab_3,
                        'nct_med_ab_4':self.nct_med_ab_4,
                        'nct_med_ab_5':self.nct_med_ab_5,
                        'nct_dspad_ab_1':self.nct_dspad_ab_1,
                        'nct_dspad_ab_2':self.nct_dspad_ab_2,
                        'nct_dspad_ab_3':self.nct_dspad_ab_3,
                        'nct_dspad_ab_4':self.nct_dspad_ab_4,
                        'nct_dspad_ab_5':self.nct_dspad_ab_5,
                        'res_ar_ab_sc':self.res_ar_ab_sc,
                        'res_agua_ab_sc':self.res_agua_ab_sc,
                        'res_osso_ab':self.res_osso_ab,
                        'res_ar_ab':self.res_ar_ab,
                        'res_agua_ab':self.res_agua_ab,
                        'res_acrilico_ab':self.res_acrilico_ab,
                        'res_polietileno_ab':self.res_polietileno_ab,
                        'obs':self.obs,
                        'nct_dspad_ab_6':self.nct_dspad_ab_6,
                        'nct_med_ab_6':self.nct_med_ab_6,
                        'nct_dspad_cr_6':self.nct_dspad_cr_6,
                        'nct_med_cr_6':self.nct_med_cr_6,
                        'nct_med_rel_ab_1':self.nct_med_rel_ab_1,
                        'nct_med_rel_ab_2':self.nct_med_rel_ab_2,
                        'nct_med_rel_ab_3':self.nct_med_rel_ab_3,
                        'nct_med_rel_ab_4':self.nct_med_rel_ab_4,
                        'nct_med_rel_ab_5':self.nct_med_rel_ab_5,
                        'nct_med_rel_ab_6':self.nct_med_rel_ab_6,
                        'nct_med_rel_cr_1':self.nct_med_rel_cr_1,
                        'nct_med_rel_cr_2':self.nct_med_rel_cr_2,
                        'nct_med_rel_cr_3':self.nct_med_rel_cr_3,
                        'nct_med_rel_cr_4':self.nct_med_rel_cr_4,
                        'nct_med_rel_cr_5':self.nct_med_rel_cr_5,
                        'nct_med_rel_cr_6':self.nct_med_rel_cr_6,
                        'resultado':self.resultado,
                        'cliente_id':self.cliente_id,
                        'equipamento_id':self.equipamento_id,
                        'user_id':self.user_id,
                        'visualized':self.visualized,
                        'print':self.print,
                        'print_date':self.print_date,
                        'print_user':self.print_user,
                        'relatorio':self.relatorio,
                        'unf_cr_a_1':self.unf_cr_a_1,
                        'unf_cr_a_2':self.unf_cr_a_2,
                        'unf_cr_a_3':self.unf_cr_a_3,
                        'unf_cr_a_4':self.unf_cr_a_4,
                        'unf_cr_a_5':self.unf_cr_a_5,
                        'ruido_cr_a_1':self.ruido_cr_a_1,
                        'ruido_cr_a_2':self.ruido_cr_a_2,
                        'ruido_cr_a_3':self.ruido_cr_a_3,
                        'ruido_cr_a_4':self.ruido_cr_a_4,
                        'ruido_cr_a_5':self.ruido_cr_a_5,
                        'unf_ab_a_1':self.unf_ab_a_1,
                        'unf_ab_a_2':self.unf_ab_a_2,
                        'unf_ab_a_3':self.unf_ab_a_3,
                        'unf_ab_a_4':self.unf_ab_a_4,
                        'unf_ab_a_5':self.unf_ab_a_5,
                        'ruido_ab_a_1':self.ruido_ab_a_1,
                        'ruido_ab_a_2':self.ruido_ab_a_2,
                        'ruido_ab_a_3':self.ruido_ab_a_3,
                        'ruido_ab_a_4':self.ruido_ab_a_4,
                        'ruido_ab_a_5':self.ruido_ab_a_5,
                        'unf_cr_ar_1':self.unf_cr_ar_1,
                        'unf_cr_ar_2':self.unf_cr_ar_2,
                        'unf_cr_ar_3':self.unf_cr_ar_3,
                        'unf_cr_ar_4':self.unf_cr_ar_4,
                        'unf_cr_ar_5':self.unf_cr_ar_5,
                        'ruido_cr_ar_1':self.ruido_cr_ar_1,
                        'ruido_cr_ar_2':self.ruido_cr_ar_2,
                        'ruido_cr_ar_3':self.ruido_cr_ar_3,
                        'ruido_cr_ar_4':self.ruido_cr_ar_4,
                        'ruido_cr_ar_5':self.ruido_cr_ar_5,
                        'unf_ab_ar_1':self.unf_ab_ar_1,
                        'unf_ab_ar_2':self.unf_ab_ar_2,
                        'unf_ab_ar_3':self.unf_ab_ar_3,
                        'unf_ab_ar_4':self.unf_ab_ar_4,
                        'unf_ab_ar_5':self.unf_ab_ar_5,
                        'ruido_ab_ar_1':self.ruido_ab_ar_1,
                        'ruido_ab_ar_2':self.ruido_ab_ar_2,
                        'ruido_ab_ar_3':self.ruido_ab_ar_3,
                        'ruido_ab_ar_4':self.ruido_ab_ar_4,
                        'ruido_ab_ar_5':self.ruido_ab_ar_5,

            }
            return resultado
```
